```python
from matplotlib.pyplot import cm
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 10]
import numpy as np

x = np.linspace(-3, 3, 1000)
y = np.linspace(-3, 4, 1000)
X, Y = np.meshgrid(x, y)

dx = np.log((Y**2 - Y + 1)/3)
dy = X**2 - Y**2

plt.streamplot(X,Y,dx, dy, color=dy-dx, density=2, cmap='cool', arrowsize=1)
plt.show()
```

```python
from matplotlib.pyplot import cm
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 10]
import numpy as np

x = np.linspace(-3, 3, 1000)
y = np.linspace(-3, 3, 1000)
X, Y = np.meshgrid(x, y)

dx = 4*X + Y
dy = 2*X + 5*Y

plt.streamplot(X,Y,dx, dy, color=dy, density=2, cmap='cool', arrowsize=1)
plt.show()
```
